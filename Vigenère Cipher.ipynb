```python
def encrypt(message, key):
    encrypted = ""                                 # for chipher text               (done)
    plain_text = message                           # variable for plain text        (done)
    plain_text_index = []                          # for index in plain text        (done)
    keys = []           #key                       # for keyword                    (done)
    keys_index = []                                # for keyword index               (done)
    formula = []                                   # ( plaintext_index + key_index ) % 26 / chipher_text_index (done)
    
    
    
    # memecah plain text sesuai panjang dari key kedalam sebuah list kedalam bentuk list-list
    split_message = [
        message[i : i + len(key)] for i in range(0, len(message), len(key))
    ]
#     print(split_message)
    
    #looping "split_message" untuk memisahkan plain text 
    for each_split in split_message:
        i = 0
        for letter in each_split:   # memisahkan plain text yang ada dalam list ke dalam huruf semula
#             print(key[i])
            number = (letter_to_index[letter] + letter_to_index[key[i]]) % len(alphabet) # rumus enkripsi
            plain_text_index.append(letter_to_index[letter]) # menambahkan index dari plain text
            keys.append(key[i])                        # menambahkan key 
            keys_index.append(letter_to_index[key[i]]) # menambahkan index dari key
            formula.append(number)                 # menambahkan index dari chipher text
            encrypted += index_to_letter[number]   # mengumpulkan tiap huruf dari chipher text kedalam variable "encrypted"
            i += 1
    
    
#     menampilkan hasil 
    def show():
        print("+====================================================================+") 
        print("| No | plain |  plain      |  kunci | kunci  | chipher     | chipher |")
        print("|    | text  |  text index |        | index  | text index  | text    |")
        print("+====================================================================+")

        a = 0
        for item in range(len(encrypted)):
            a +=1


            print(f"| {a}  |\t{plain_text[item]}    |\t{plain_text_index[item]}\t   |\t{keys[item]}   |{keys_index[item]}\t     |\t{formula[item]}\t   |   {encrypted[item]}     |")


        print("+====================================================================+")
    
    #memanggil fungsi show
    show()
    print("Plain text  \t:", plain_text )
    print("Cipher Text\t:", encrypted)
    
    decrypt(encrypted, key)
    
# encrypt("cupu","banana")

```

```python
def decrypt(cipher, key):
    print()
    print("Dekripsi".upper())
    decrypted = ""
    cipher_text = cipher
    cipher_text_index = []
    keys = []
    keys_index = []
    formula = []
    plaintext_index = []
    
    # memecah pesan yang telah di enkripsi menjadi sebuah list yang terdapat list-list dengan panjang masing-masing
    # sesuai panjang kuncinya
    split_encrypted = [
        cipher[i : i + len(key)] for i in range(0, len(cipher), len(key))
    ]
    # memecah variable list "split_encrypted" ke dalam variable "each_split" menjadi sebuah list-list
    #yang memiliki panjang sama dengan kunci
    for each_split in split_encrypted:
        i = 0                      # karena dipecah berdasarkan list2 kecil, maka looping variable "i" hanya akan sampai panjang kuncinya, kemudian mengulang lagi dari nol
        for letter in each_split:
           
            number = (letter_to_index[letter] - letter_to_index[key[i]]) % len(alphabet)
            
            cipher_text_index.append(letter_to_index[letter])                         # menambahkan index dari cipher text
            keys.append(key[i])                                                       # menambahkan isi dari key
            keys_index.append(letter_to_index[key[i]])                                # menambahkan indeks dari key
            formula.append(number)
            
            decrypted += index_to_letter[number]
            i += 1

    
    
    def show():
        print("+====================================================================+") 
        print("| No | cipher|  cipher     |  kunci | kunci  | plain       | plain   |")
        print("|    | text  |  text index |        | index  | text index  | text    |")
        print("+====================================================================+")

        a = 0
        for item in range(len(decrypted)):
            a +=1


            print(f"| {a}  |\t{cipher_text[item]}    |\t{cipher_text_index[item]}\t   |\t{keys[item]}   |{keys_index[item]}\t     |\t{formula[item]}\t   |   {decrypted[item]}     |")


        print("+====================================================================+")
    
    #memanggil fungsi show
    show()
    print("Cipher Text\t:", cipher_text )
    print("Plain Text \t:", decrypted)
        
    
#     print("cipher text: ", cipher_text)    
#     print("index cipher text: ", cipher_text_index)
#     print("keys:", keys )
#     print("index key: ", keys_index )
#     print("index plaintext:", formula)
#     print("hasil:", decrypted )

# decrypt('glvdniqsdaf', 'gans')
```
